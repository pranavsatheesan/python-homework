```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""


# @TODO: Import libraries
import csv
from pathlib import Path
import os

#this is to create a DEEPCOPY of teh dictionary in the end
import copy

# to see current directory --- uncomment to check
#Path.cwd()

# ********Ensure to CHANGE this Path to the one in your station************
os.chdir('C:/Users/Sobha/Desktop/Uoft/Homework/python-homework/PyRamen')

menu = []
sales = []

# import the menu file
imported_file1 = open('menu_data.csv', 'r')
#read the csv file 
imported_file2 = csv.reader(imported_file1, delimiter = ',')

csv_header1 = next(imported_file2)
#debug check
#print(csv_header)

#transfer the imported file to menu without the header
menu = list(imported_file2)
# debug check to see if file is transferred to menu correctly. uncomment to check
#print(menu)


# import the sales file
imported_file3 = open('sales_data.csv', 'r')
#read the csv file version of sales
imported_file4 = csv.reader(imported_file3, delimiter = ',')

#transfer the header of the file
csv_header2 = next(imported_file4)
# debug check print(csv_header2)

#transfer data to sales list
sales = list(imported_file4)
#print(sales) debug check


quantity = []
menu_item = []
food_type =[]

#get all UNIQUE values of ramen types into list food_type
for i in sales:
    if i[4] not in food_type:
        food_type.append(i[4])

quantity_total = 0
quantity_total_list = []

#get total quantity of each food type and store in quantity_total_list
#quantitry_total_list will have [food type, total, food type, total....etc]
for i in food_type:
    for i2 in sales:
        if i2[4] == i:
            quantity_total = quantity_total + int(i2[3])
    quantity_total_list.append(i)
    quantity_total_list.append(quantity_total)
    quantity_total = 0
#print(quantity_total_list) debug check

#create list of lists where format is [(food type, quanity), (food type, quantity)]
b = zip(*[iter(quantity_total_list)]*2)
quantity_master = list(b)
#print(quantity_master) debug check
#print(quantity_master)
del i
del i2

b = []
# get cost revenue and profit for each ramen
for i in quantity_master:
    for i2 in menu:
        if i[0] == i2[0]:
            total_cost = int(i2[4]) * int(i[1])
            total_revenue = int(i2[3]) * int(i[1])
            total_profit = total_revenue - total_cost
            #create list of format [ramen, quantity, revenue, cost, profit....ramen, quantity, revenue, cost, profit]
            b.append(i[0])
            b.append(i[1])
            b.append(total_revenue)
            b.append(total_cost)
            b.append(total_profit)
#print(b) debug check

#create final results but in list format insteatd of dictionary
final_format = zip(*[iter(b)]*5)
final_format_list = list(final_format)

#print(final_format_list,"\n\n")

#initialize the sub dictionary
report2 = {
"01-count": 0,
"02-revenue": 0,
"03-cogs": 0,
"04-profit": 0,
}

#create dictionary structure for the final output
report = {}
for i in food_type:
    #Create deepcopy, else it will fail
    report[i] = copy.deepcopy(report2)
#print(report)

#create deep copy again else all we get is shallw copies which will effectively point to the same location as the original dictionary
new_dict_deepcopy = copy.deepcopy(report)

#debug test to check if deepcopy is working
#new_dict_deepcopy['tonkotsu ramen']['01-count'] = 89
#print(new_dict_deepcopy)

for key,value in new_dict_deepcopy.items():
    for i in final_format_list:
        if key == i[0]:
            new_dict_deepcopy[key]['01-count'] = i[1]
            new_dict_deepcopy[key]['02-revenue'] = i[2]
            new_dict_deepcopy[key]['03-cogs'] = i[3]
            new_dict_deepcopy[key]['04-profit'] = i[4]
            
#this is thefinal dictionary with the result deebug check
#print(new_dict_deepcopy)
            
del key
del value

#print in the required format on screen
for key,value in new_dict_deepcopy.items():
    print(key, ':', value)

#Create a Text file to output results OR replace if the text file already exists
f = open("results_pyramen_pranav.txt", "w")
f.close()
    
#open the created text file for writing results
f =open("results_pyramen_pranav.txt", "a")

del key
del value

#print in the required format to text file
for key,value in new_dict_deepcopy.items():
    print(key, ':', value, file = f)

# close text file
f.close()
    
#close the imported file
imported_file1.close()
imported_file3.close()

```
